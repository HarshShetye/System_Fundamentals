```python
def LRU(pages, frames):
    page_faults = 0
    page_frames = [None] * frames
    page_usage = [-1] * frames
    for i, page in enumerate(pages):
        if page not in page_frames:
            if None in page_frames:
                page_frames[page_frames.index(None)] = page
            else:
                page_frames[page_usage.index(min(page_usage))] = page
            page_faults += 1
        page_usage[page_frames.index(page)] = i
        print(page_frames)
    hit_ratio = 1 - (page_faults / len(pages))
    return page_faults, hit_ratio


pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3, 2]
frames = 4
faults, hit_ratio = LRU(pages, frames)
print(f"Number of Page Faults using LRU: {faults}")
print(f"Hit Ratio using LRU {hit_ratio}")
```

```python
def Optimal(pages, frames):
    page_faults = 0
    page_frames = [None] * frames
    for i, page in enumerate(pages):
        if page not in page_frames:
            if None in page_frames:
                page_frames[page_frames.index(None)] = page
            else:
                replace_page = max(page_frames, key=lambda x: pages[i+1:].index(x) if x in pages[i+1:] else float('inf'))
                page_frames[page_frames.index(replace_page)] = page
            page_faults += 1
        print(page_frames)
    hit_ratio = 1 - (page_faults / len(pages))
    return page_faults, hit_ratio


pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3, 2]
frames = 4
faults, hit_ratio = Optimal(pages, frames)
print(f"Number of Page Faults using Optimal: {faults}")
print(f"Hit Ratio using Optimal {hit_ratio}")
```

```python
def FIFO(pages, frames):
    page_faults = 0
    page_frames = [None] * frames
    frame_pointer = 0

    for page in pages:
        if page not in page_frames:
            page_frames[frame_pointer] = page
            frame_pointer = (frame_pointer + 1) % frames
            page_faults += 1
        print(page_frames)

    hit_ratio = 1 - (page_faults / len(pages))
    return page_faults, hit_ratio


pages_string = [1, 3, 0, 3, 5, 6]
page_frames = 3
faults, hitratio = FIFO(pages_string, page_frames)
print(f"Number of Page Faults using FIFO: {faults}")
print(f"Hit Ratio using FIFO {hitratio}")
```

```python
def fcfs(processes):
    n = len(processes)
    wt = [0] * n
    tat = [0] * n
    ct = [0] * n

    ct[0] = processes[0][1]  # completion time of first process is its burst time
    # turn around time of first process is its completion time - arrival time
    tat[0] = ct[0] - processes[0][0]
    # waiting time of first process is its turn around time - burst time
    wt[0] = tat[0] - processes[0][1]
    gantt_chart = f"0->{'P1'}->{processes[0][1]}"

    for i in range(1, n):
        # completion time of current process is completion time of previous process + current process's burst time
        ct[i] = ct[i-1] + processes[i][1]
        # turn around time of current process is its completion time - arrival time
        tat[i] = ct[i] - processes[i][0]
        # waiting time of current process is its turn around time - burst time
        wt[i] = tat[i] - processes[i][1]
        gantt_chart += f"->{'P'+str(i+1)}->{ct[i]}"

    total_wt = sum(wt)
    total_tat = sum(tat)
    avg_wt = total_wt / n
    avg_tat = total_tat / n
    print("TABLE")
    print("Process\tAT\tBT\tCT\tTAT\tWT")
    for i in range(n):
        print(
            f"P{i+1}\t{processes[i][0]}\t{processes[i][1]}\t{ct[i]}\t{tat[i]}\t{wt[i]}")
    print("\nAverage Turn Around Time:", avg_tat,
          "Average Waiting Time:", avg_wt)
    print("\nGantt Chart:")
    print(gantt_chart)


def sjf(processes):
    n = len(processes)
    wt = [0] * n
    tat = [0] * n
    ct = [0] * n
    rt = [0] * n  # remaining time
    done = [False] * n  # keep track of completed processes

    for i in range(n):
        # remaining time of each process is initially its burst time
        rt[i] = processes[i][1]

    current_time = 0
    total_processes_completed = 0
    gantt_chart = ""

    while total_processes_completed < n:
        shortest = -1
        for i in range(n):
            if not done[i] and (shortest == -1 or rt[i] < rt[shortest]) and processes[i][0] <= current_time:
                shortest = i

        if shortest == -1:
            # no process can be executed at current time, so increment time by 1
            current_time += 1
            continue

        # completion time of current process
        ct[shortest] = current_time + rt[shortest]
        # turn around time of current process
        tat[shortest] = ct[shortest] - processes[shortest][0]
        # waiting time of current process
        wt[shortest] = tat[shortest] - processes[shortest][1]
        done[shortest] = True  # mark current process as completed
        total_processes_completed += 1

        gantt_chart += f"->{'P'+str(shortest+1)}->{ct[shortest]}"
        # update current time to completion time of current process
        current_time = ct[shortest]

    total_wt = sum(wt)
    total_tat = sum(tat)
    avg_wt = total_wt / n
    avg_tat = total_tat / n

    print("TABLE")
    print("Process\tAT\tBT\tCT\tTAT\tWT")
    for i in range(n):
        print(
            f"P{i+1}\t{processes[i][0]}\t{processes[i][1]}\t{ct[i]}\t{tat[i]}\t{wt[i]}")
    print("\nAverage Turn Around Time:", avg_tat,
          "Average Waiting Time:", avg_wt)
    print("\nGantt Chart:")
    print('0'+gantt_chart)


def priority_scheduling(processes):
    n = len(processes)
    # Sort the processes by their priority (in ascending order)
    processes.sort(key=lambda x: x[2])

    # Initialize variables for calculating CT, TAT, and WT
    ct = [0] * n
    tat = [0] * n
    wt = [0] * n
    total_tat = 0
    total_wt = 0

    # Calculate CT, TAT, and WT for each process
    for i in range(n):
        # Set the CT for the first process as the AT + BT
        if i == 0:
            ct[i] = processes[i][0] + processes[i][1]
        else:
            # For subsequent processes, CT is the maximum of their AT and the CT of the previous process
            ct[i] = max(processes[i][0], ct[i-1]) + processes[i][1]
        # Calculate TAT and WT for each process
        tat[i] = ct[i] - processes[i][0]
        wt[i] = tat[i] - processes[i][1]
        total_tat += tat[i]
        total_wt += wt[i]

    # Calculate average TAT and WT
    avg_tat = total_tat / n
    avg_wt = total_wt / n

    # Print the Gantt Chart and table
    print("Gantt Chart")
    curr_time = processes[0][0]
    for i in range(n):
        print(curr_time, "->", "P" + str(processes[i][3]), end=" -> ")
        curr_time += processes[i][1]
    print(curr_time)

    print("\nTABLE")
    print("Process\tAT\tBT\tPriority\tCT\tTAT\tWT")
    for i in range(n):
        print("P" + str(processes[i][3]), "\t", processes[i][0], "\t", processes[i]
              [1], "\t", processes[i][2], "\t\t", ct[i], "\t", tat[i], "\t", wt[i])

    # Print average TAT and WT
    print("\nAverage Turn Around Time:", avg_tat,
          "Average Waiting Time:", avg_wt)


print('Harsh Shetye_60009210068\n')
n = int(input("Enter the no. of processes: "))
processes = []
for i in range(n):
    at, bt, priority = map(int, input(
        f"Enter the arrival time, burst time, and priority for process P{i+1}: ").split())
    processes.append((at, bt, priority, i+1))

choice = int(input(
    'Select amongst the following Non-Preemptive Scheduling Algorithm\n1. FCFS\n2. SJF\n3. Priority\n'))
if choice == 1:
    print("FCFS\n")
    fcfs(processes)
elif choice == 2:
    print("SJF\n")
    sjf(processes)
elif choice == 3:
    print("PRIORITY\n")
    priority_scheduling(processes)
else:
    print("Invalid Choice")
```

```python
# Harsh Shetye_60009210068_A2_SF_Memory Allocation
def firstFit(blockSize, processSize):
    m = len(blockSize)
    n = len(processSize)
    allocation = [-1] * n
    for i in range(n):
        for j in range(m):
            if blockSize[j] >= processSize[i]:
                allocation[i] = j
                blockSize[j] -= processSize[i]
                break
    print(" Process No. | Process Size  |   Block no.")
    for i in range(n):
        print("     ", i + 1, "          ",
              processSize[i], "          ", end=" ")
        if allocation[i] != -1:
            print(allocation[i] + 1)
        else:
            print("Not Allocated")


def bestFit(blockSize, processSize):
    m = len(blockSize)
    n = len(processSize)
    allocation = [-1] * n
    for i in range(n):
        Idx = -1
        for j in range(m):
            if blockSize[j] >= processSize[i]:
                if Idx == -1:
                    Idx = j
                elif blockSize[Idx] > blockSize[j]:
                    Idx = j
        if Idx != -1:
            allocation[i] = Idx
            blockSize[Idx] -= processSize[i]
    print(" Process No. | Process Size  |   Block no.")
    for i in range(n):
        print("     ", i + 1, "          ",
              processSize[i], "          ", end=" ")
        if allocation[i] != -1:
            print(allocation[i] + 1)
        else:
            print("Not Allocated")


def worstFit(blockSize, processSize):
    m = len(blockSize)
    n = len(processSize)
    allocation = [-1] * n
    for i in range(n):
        Idx = -1
        for j in range(m):
            if blockSize[j] >= processSize[i]:
                if Idx == -1:
                    Idx = j
                elif blockSize[Idx] < blockSize[j]:
                    Idx = j
        if Idx != -1:
            allocation[i] = Idx
            blockSize[Idx] -= processSize[i]
    print(" Process No. | Process Size  |   Block no.")
    for i in range(n):
        print("     ", i + 1, "          ",
              processSize[i], "          ", end=" ")
        if allocation[i] != -1:
            print(allocation[i] + 1)
        else:
            print("Not Allocated")


blockSize = list(
    map(int, input("Enter block sizes separated by space: ").split()))
processSize = list(
    map(int, input("Enter process sizes separated by space: ").split()))
choice = int(input(
    'Select amongst the following Contiguous Memory Allocation Policy\n1. Best Fit\n2. First Fit\n3. Worst Fit\n'))
if choice == 1:
    print("Best Fit\n")
    bestFit(blockSize, processSize)
elif choice == 2:
    print("First Fit\n")
    firstFit(blockSize, processSize)
elif choice == 3:
    print("Worst Fit\n")
    worstFit(blockSize, processSize)
else:
    print("Invalid Choice")
```

```python
def bankers_algorithm(num_resources, max_resources, alloc_matrix, max_matrix):
    available = [max_resources[i] - sum(alloc_matrix[j][i]
                                        for j in range(num_processes)) for i in range(num_resources)]
    finished = [False] * num_processes
    safe_sequence = []
    while False in finished:
        safe = False
        for i in range(num_processes):
            if not finished[i] and all(max_matrix[i][j] - alloc_matrix[i][j] <= available[j] for j in range(num_resources)):
                available = [available[j] + alloc_matrix[i][j]
                             for j in range(num_resources)]
                safe_sequence.append(i)
                finished[i] = True
                safe = True
                break
        if not safe:
            return "Unsafe state"

    return "< " + "->".join([f"P[{i}]" for i in safe_sequence]) + " >"


num_resources = 3
max_resources = [10, 7, 7]
num_processes = 3
alloc_matrix = [
    [3, 2, 1],
    [1, 1, 2],
    [4, 1, 2]
]
max_matrix = [
    [4, 4, 4],
    [3, 4, 5],
    [5, 2, 4]
]
print(bankers_algorithm(num_resources, max_resources, alloc_matrix, max_matrix))
```

```python
import matplotlib.pyplot as plt


def cscan(start, requests, track_limit, seek_time_per_cylinder):
    sequence = sorted(requests)
    sequence.remove(start) if start in sequence else None
    sequence = [start] + [x for x in sequence if x > start] + [track_limit] + [0] + [x for x in (sequence) if x < start]
    movement = sum(abs(sequence[i+1] - sequence[i]) for i in range(len(sequence)-1))
    return sequence, movement, seek_time_per_cylinder * movement


start = 20
seek_time_per_cylinder = 6
requests = [10, 22, 20, 2, 40, 6, 38]
track_limit = 49
cscan_sequence, total_movement, total_seek_time = cscan(
    start, requests, track_limit, seek_time_per_cylinder)

print("CSCAN sequence:", cscan_sequence)
print("CSCAN total movement:", total_movement)
print("CSCAN total seek time:", total_seek_time, " msec")

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(cscan_sequence, range(len(cscan_sequence)),
        'm', marker='o', linewidth=2, markersize=8)
ax.set_xlabel("Requests")
ax.set_title("CSCAN")
for i, req in enumerate(cscan_sequence):
    ax.text(req, i, req, size=15)
ax.xaxis.tick_top()
ax.invert_yaxis()
plt.show()
```

```python
import matplotlib.pyplot as plt


def scan(start, requests, track_limit, seek_time_per_cylinder):
    sequence = sorted(requests)
    sequence.remove(start) if start in sequence else None
    sequence = [start] + [x for x in sequence if x > start] + \
        [track_limit] + [x for x in reversed(sequence) if x < start]
    movement = sum(abs(sequence[i+1] - sequence[i])
                   for i in range(len(sequence)-1))
    return sequence, movement, seek_time_per_cylinder * movement


start = 20
seek_time_per_cylinder = 6
requests = [10, 22, 20, 2, 40, 6, 38]
track_limit = 49
scan_sequence, total_movement, total_seek_time = scan(
    start, requests, track_limit, seek_time_per_cylinder)

print("SCAN sequence:", scan_sequence)
print("SCAN total movement:", total_movement)
print("SCAN total seek time:", total_seek_time, " msec")

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(scan_sequence, range(len(scan_sequence)),
        'g', marker='o', linewidth=2, markersize=8)
ax.set_xlabel("Requests")
ax.set_title("ELEVATOR (SCAN)")
for i, req in enumerate(scan_sequence):
    ax.text(req, i, req, size=15)
ax.xaxis.tick_top()
ax.invert_yaxis()
plt.show()
```

```python
import matplotlib.pyplot as plt


def FCFS(start, requests, track_limit, seek_time_per_cylinder):
    sequence = [start] + requests
    movement = sum(abs(sequence[i+1] - sequence[i])
                   for i in range(len(sequence)-1))
    return sequence, movement, seek_time_per_cylinder * movement


start = 20
seek_time_per_cylinder = 6
requests = [10, 22, 20, 2, 40, 6, 38]
track_limit = 49
fsfs_sequence, total_movement, total_seek_time = FCFS(
    start, requests, track_limit, seek_time_per_cylinder)

print("FCFS sequence:", fsfs_sequence)
print("FCFS total movement:", total_movement)
print("FCFS total seek time:", total_seek_time, " msec")

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(fsfs_sequence, range(len(fsfs_sequence)),
        'r', marker='o', linewidth=2, markersize=8)
ax.set_xlabel("Requests")
ax.set_title("FCFS")
for i, req in enumerate(fsfs_sequence):
    ax.text(req, i, req, size=15)
ax.xaxis.tick_top()
ax.invert_yaxis()
plt.show()
```

```python
import matplotlib.pyplot as plt


def SSTF(start, requests, track_limit, seek_time_per_cylinder):
    current = start
    total_seek_time = 0
    visited_sequence = []
    while requests:
        distances = [abs(request - current) for request in requests]
        closest_index = distances.index(min(distances))
        closest_request = requests.pop(closest_index)
        total_seek_time += distances[closest_index] * seek_time_per_cylinder
        visited_sequence.append(closest_request)
        current = closest_request
    return total_seek_time, visited_sequence


start = 20
seek_time_per_cylinder = 6
requests = [10, 22, 20, 2, 40, 6, 38]
track_limit = 49
total_seek_time, sstf_sequence = SSTF(
    start, requests, track_limit, seek_time_per_cylinder)

print("SSTF sequence:", sstf_sequence)
print("SSTF total movement:", total_seek_time/seek_time_per_cylinder)
print("SSTF total seek time:", total_seek_time, " msec")

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(sstf_sequence, range(len(sstf_sequence)),
        'y', marker='o', linewidth=2, markersize=8)
ax.set_xlabel("Requests")
ax.set_title("SSTF")
for i, req in enumerate(sstf_sequence):
    ax.text(req, i, req, size=15)
ax.xaxis.tick_top()
ax.invert_yaxis()
plt.show()
```

```python
def LRU(pages, frames):
    page_faults, page_hits = 0, 0
    page_frames = [None] * frames
    page_usage = [-1] * frames
    for i, page in enumerate(pages):
        if page not in page_frames:
            if None in page_frames:
                page_frames[page_frames.index(None)] = page
            else:
                page_frames[page_usage.index(min(page_usage))] = page
            page_faults += 1
            print(f"Address:{page}  Fault:{page_faults}", page_frames)
        else:
            page_hits += 1
            print(f"Address:{page}  Hits:{page_hits}", page_frames)
        page_usage[page_frames.index(page)] = i
    hit_ratio = 1 - (page_faults / len(pages))
    return page_faults, hit_ratio


pages = ['0x80000000', '0x80000004', '0x80000008', '0x8000000C', '0x00000020', '0x80000010',
         '0x80000014', '0x80000018', '0x8000000C', '0x00000024', '0x80000010', '0x80000014',
         '0x80000018', '0x8000000C', '0x00000028', '0x80000010', '0x80000014', '0x80000018',
         '0x8000000C', '0x0000002C', '0x80000010', '0x80000014', '0x80000018', '0x8000000C',
         '0x00000030', '0x80000010', '0x80000014', '0x80000018', '0x8000000C', '0x00000034',
         '0x80000010', '0x80000014', '0x80000018', '0x8000000C', '0x00000038', '0x80000010',
         '0x80000014', '0x80000018', '0x8000000C', '0x0000003C', '0x80000010', '0x80000014',
         '0x80000018', '0x8000000C', '0x00000040', '0x80000010', '0x80000014', '0x80000018',
         '0x8000000C', '0x00000044', '0x80000010', '0x80000014', '0x80000018', '0x8000000C',
         '0x00000048', '0x80000010', '0x80000014', '0x80000018', '0x8000000C', '0x0000004C',
         '0x80000010', '0x80000014', '0x80000018', '0x8000000C', '0x00000050', '0x80000010',
         '0x80000014', '0x80000018', '0x8000000C', '0x00000054', '0x80000010', '0x80000014',
         '0x80000018', '0x8000000C', '0x00000058', '0x80000010', '0x80000014', '0x80000018',
         '0x8000000C', '0x0000005C', '0x80000010', '0x80000014', '0x80000018', '0x8000000C',
         '0x00000060', '0x80000010', '0x80000014', '0x80000018', '0x8000000C', '0x00000064',
         '0x80000010', '0x80000014', '0x80000018', '0x8000000C', '0x00000068', '0x80000010',
         '0x80000014', '0x80000018', '0x8000000C', '0x0000006C', '0x80000010', '0x80000014',
         '0x80000018']
frames = 8
faults, hit_ratio = LRU(pages, frames)
print(f"Number of Page Faults using LRU: {faults}")
print(f"Hit Ratio using LRU {hit_ratio}")
print(len(pages)-faults)
```
